```python
import gradio as gr
from transformers import pipeline

# Model from the tutorial
MODEL_NAME = "linkanjarad/mobilenet_v2_1.0_224-plant-disease-identification"

clf = pipeline(
    "image-classification",
    model=MODEL_NAME
)

def analyze_plant(image, temp, humidity, soil):
    preds = clf(image)
    top = preds[0]

    alerts = []
    advice = []

    if soil < 25:
        alerts.append("Low soil moisture")
        advice.append("Recommendation: irrigate / water the plant")

    if humidity > 80:
        alerts.append("High humidity")
        advice.append("Recommendation: improve ventilation (reduce fungal risk)")

    if not alerts:
        alerts.append("Status looks normal")

    return (
        f"Detected disease: {top['label']} ({top['score']:.2%})",
        "\n".join(alerts),
        "\n".join(advice)
    )

gr.Interface(
    fn=analyze_plant,
    inputs=[
        gr.Image(type="filepath", label="Plant Image"),
        gr.Slider(0, 45, label="Temperature (Â°C)"),
        gr.Slider(0, 100, label="Humidity (%)"),
        gr.Slider(0, 100, label="Soil Moisture (%)"),
    ],
    outputs=[
        gr.Textbox(label="Diagnosis"),
        gr.Textbox(label="Alerts"),
        gr.Textbox(label="Recommendations"),
    ],
    title="ðŸŒ± Smart Plant Monitoring Dashboard",
    description="Plant disease detection using Hugging Face + basic care recommendations"
).launch()
```
